# Read Data

```python
fp = pd.read_csv('./data/MIR_Fruit_purees.csv')
```

```python
fp.sample(10)
```

```python
fpt = fp.iloc[:,1:].transpose()
fpt.sample(10)
```

Next, we need to extract the label:
- 1 for strawberry
- 0 for other

```python
fpt['strawberry'] = fpt.index
fpt['strawberry'] = ~fpt.strawberry.str.contains('NON')
fpt['strawberry'] = [1  if x else 0 for x in fpt.strawberry]
fpt.strawberry.value_counts()
```

```python
fpt.sample(5)
```

Plot the data...

```python
x = fpt.iloc[:,:-1]
y = fpt.strawberry
x.sample(10).transpose().plot(figsize=(16,10));
```

# Simple Modeling
Let's start by creating a simple Naive Bayes model...

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
nb = GaussianNB()
cross_val_score(nb, x, y)
```

```python
y.value_counts()
```

## Confusion Matrix

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=3)
nb.fit(x_train, y_train)
preds = nb.predict(x_test)
cm = confusion_matrix(preds, y_test)
print(classification_report(preds, y_test))
print(cm)
```

# Dimensionality Reduction
The model above uses 235 dimensions. Looking at the graph every dimension is highly correlated to it's neighbors. So, do we really need so many dimensions to express all the information?

**Meet the PCA!**

```python
# Generate example data

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager
from sklearn import svm
from sklearn.ensemble import IsolationForest

df = pd.DataFrame(np.random.randn(250, 2))
df[1] = df[0] + df[1] * 0.7

fix, ax = plt.subplots(figsize=(6,6))
plt.scatter(df[0], df[1], color='grey')
plt.xlabel('x_1')
plt.ylabel('x_2');
```

Most of the information in this data set is along a linear axis along the diagonal, so if we can somehow extract the direction of the main axis, we can express a huge part of the information with the position of the projection of the data onto this line.

This is what the PCA can do for you!

The PCA creates an axis transformation. It creates new axes, such that:
- all new axes are orthogonal
- each axis describes as much variance as possible

Have a look at the result:

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver='full')
fix, ax = plt.subplots(figsize=(6,6))
plt.scatter(df[0], df[1], color='grey')

pca.fit(df)
pca_score = pca.explained_variance_ratio_
pca_comp = (pca.components_)
plt.plot([0,pca_comp[0,0] * pca_score[0] / pca_score.min()],[0,pca_comp[0,1] * pca_score[0] / pca_score.min()])
plt.plot([0,pca_comp[1,0] * pca_score[1] / pca_score.min()],[0,pca_comp[1,1] * pca_score[1] / pca_score.min()])
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.xlim(-4,3)
plt.ylim(-4,3)
pca_comp
```

And here is the data in the transformed space:

```python
df_pca = pd.DataFrame(pca.transform(df))
fig, ax = plt.subplots(figsize=(6,6))
plt.scatter(df_pca[0], df_pca[1], color='grey')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
plt.xlabel('pc_1')
plt.ylabel('pc_2')
plt.plot([0,1], [0,0])
plt.plot([0,0], [0,1])
```

Now we only need the first principal component (pc_1) to express most of the variance in the data. How much?

```python
pca.explained_variance_ratio_
```

## How does it work?
The principal components produced by the PCA are the Eigenvectors of the covariance matrix.

The explained variance of each component corresponds to the Eigenvalues of the covariance matrix.

# Modeling with PCA

The PCA is availabe as `sklearn.decomposition.PCA`.

When you use it, it's usually a good idea to standardize the data before the PCA. Otherwise, the features with the biggest scale are overweighted for their contribution to the variance.

Usually, the PCA is used to reduce the dimensionality of the data. Usually, we specify that we want to keep a certain amount of explained variance. We then add principal components until the cumulative explained variance reaches the desired value.

Let's take a look at our strawberry puree:

Imports:

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
pca = PCA(svd_solver='full')
pca.fit(x)
```

Let's plot the cumulative explained variance:

```python
plt.plot(pca.explained_variance_ratio_.cumsum());
```

The first few components explain almost all the variance. Let's zoom in to the first 10 components:

```python
plt.plot(pca.explained_variance_ratio_[:10].cumsum());
```

We can tell the PCA to keep only the as many components as necessary to keep a certain amount of variance:

```python
import seaborn as sns

pca = PCA(svd_solver='full', n_components=0.9)
pca.fit(x)
plt.plot(pca.explained_variance_ratio_.cumsum())

pca_df = pd.DataFrame(pca.transform(x))
pca_df['l'] = y.values
sns.pairplot(data=pca_df, hue='l');
```

When using the PCA in a validation, you may only use the training data to fit it. If you would use the full data set, information from the test data would leak into the model.

So here's a pipeline with standard scaler, pca, naive bayes, and cross validation:

```python
pipe = make_pipeline(StandardScaler(), pca, nb)
pipe.fit(x, y)

# print the results and the plot PCA
plt.plot(pca.explained_variance_ratio_.cumsum())
print('with PCA: ', cross_val_score(pipe, x, y))
print('without PCA: ', cross_val_score(nb, x, y));
```

So, with only 4 dimensions our accuracy is better than when we use all 235 original dimensions!
