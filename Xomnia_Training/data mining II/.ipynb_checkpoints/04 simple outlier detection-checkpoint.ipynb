# The Boxplot
The boxplot shows for each attribute:
- the median
- the quartiles (upper and lower edge of the box)
- the upper end of the whisker is min(median + 1.5 * IQR, max(datavalue))
- lower end accordingly
- IQR is the inter-quartile range, i.e., the difference between upper quartile and lower quartile
- if there are any datapoints further away from the median than 1.5 IQR, they are plotted as outliers

Practical advice: if the scale of the data is very different, it may make sense to standardize them.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_color_codes()

fig = plt.figure(figsize=(16,10))
ax = sns.boxplot(data=df)
```

# Filter Outliers based on Standard Deviation

If the data is normally distributed, then it is within a certain probability within the average +- a multiple of the standard deviation.

 *(source: http://www.biologyforlife.com/standard-deviation.html)*

So we could assume that all data that is further away than $x$ standard deviations from the mean, it is an outlier.

```python
# standardize all data
std_df = pd.DataFrame(stats.zscore(df))
```
